```python
selected_crimes = ["Simple Assualt", "Intimidation", "Theft from Building", "Theft from Motor Vehicle","Motor Vehicle Theft", "Aggravated Assault", "Burglary/Breaking & Entering", "Robbery","Forcible Rape","Kidnapping/Abduction", "Purse-snatching","Murder & Nonnegligent Manslaughter","Pocket-Picking",]
worst_crimes = crime_activity[crime_activity["NIBRS_Offense_Type"].isin(selected_crimes)]
neighborhood_crimes2 = worst_crimes['Neighborhood'].value_counts()
highest2 = neighborhood_crimes2.tail(25)
print(highest2)
colors = ["tomato","darkorange","olive","springgreen","royalblue","mediumorchid"]
bargraph = highest2.plot(kind='bar', color = colors, title = "Top 25 Neighboorhoods with lowest crime rate Adjusted", figsize = (20,20))
plt.xlabel("Neighborhood")
plt.ylabel("Total Crimes")
```
