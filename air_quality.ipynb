Air Quality

```python
import pandas as pd
```

```python
air_quality_df = pd.read_csv('air_quality.csv')
```

```python
last_2_years = air_quality_df[63283:]
```

```python
last_2_years
```

```python
# combine the data for Lawrenceville and Lawrenceville 2, and Liberty and Liberty 2
last_2_years['site'] = last_2_years['site'].replace({'Lawrenceville 2': 'Lawrenceville', 'Liberty 2': 'Liberty'})
# rank the neighborhoods by their average air quality index (AQI)
neighborhood_aqi = last_2_years.groupby('site')['index_value'].mean().reset_index()
neighborhood_aqi = neighborhood_aqi.sort_values(by='index_value')
neighborhood_aqi
```

```python
# rank by health advisories
advisory_counts = last_2_years.groupby('site')['health_advisory'].count().reset_index()
advisory_counts = advisory_counts.sort_values(by='health_advisory', ascending=True)
advisory_counts
```
